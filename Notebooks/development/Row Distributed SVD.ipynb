### KNN-based sd estimatation

```python
import copy
import itertools
import time
import os

import luigi
import dill

from luigi_utils.plots import plot_CCCs
from empirical_privacy.row_distributed_svd import gen_SVD_CCCs_for_multiple_docs, \
svd_dataset_settings
from empirical_privacy.config import LUIGI_COMPLETED_TARGETS_DIR
import dataset_utils.common as cm
```

```python
CCCs = []

for dataset in ['20NG', 'ml-1m']:
    for trials in range(5,11):
        for part_fraction in [0.01, 0.1, 0.3]:
            for n_max in [2**8, 2**9, 2**10, 2**11]:
                start = time.time()
                print(f'dataset={dataset} trials={trials} pf={part_fraction} n_max={n_max} ')
                ds = svd_dataset_settings(dataset_name=dataset, part_fraction=part_fraction)
                _CCCs = gen_SVD_CCCs_for_multiple_docs(n_max=n_max,
                                                      validation_set_size=512,
                                                      n_docs=5,
                                                      n_trials_per_training_set_size=trials,
                                                      dataset_settings=ds)
                luigi.build(_CCCs, workers=3, log_level='ERROR')
                print(f'\tCompleted {time.time()-start}s')
                CCCs += _CCCs
```

```python
%%time
luigi.build(CCCs, workers=3, log_level='ERROR')
```
